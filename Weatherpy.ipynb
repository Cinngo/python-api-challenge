```python
import statsmodels.api as sm
df.rename(columns={'Max Temp':'temp'},inplace=True)
y=df.temp
x=df.Lat
x=sm.add_constant(x)
model=sm.OLS(y,x)
model_fit=model.fit()
model_fit.summary()
```
